```python
import pandas as pd
import datetime
import plotly.graph_objects as go
import plotly.express as px
from plotly.subplots import make_subplots
import yfinance as yf
```

```python
# fazer web scraping para pegar os dados do yahoo finance e baixar os dados ou usar algum dado para consultar as ações.
viia = yf.Ticker("VIIA3.SA")
viia  = viia.history(period="5y")


mglu = yf.Ticker("MGLU3.SA")
mglu  = mglu.history(period="5y")
```

```python
result = mglu.merge(viia, how='left', on='Date', sort=True).reset_index(drop = False)
result.isnull().sum()
result.head()
```

```python
AnoInicial = 2017
Anos = []
Valores = [6.20, 15.0]

for i in range(0,7):
    AnoInicial += 1
    if AnoInicial == 2023:
        break
    datas = str(AnoInicial) + '-' + '01' + '-' + '01'
    Anos.append(datas)

fig = px.line(result, x='Date', y=['Close_x','Close_y'])

fig.add_vline(x='2022-08-26', line_width=2, line_color="green")

for l in range(5):
    fig.add_vline(x=Anos[l], line_width=1, line_dash="dash", line_color="black")
    
for v in range(2):    
    fig.add_hline(y=Valores[v], line_color="blue")
    
fig.update_xaxes(rangeslider_visible=True)
fig.show()
```
